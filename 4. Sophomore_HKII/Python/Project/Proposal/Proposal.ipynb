### Phần III: Kế hoạch phân tích dữ liệu

**1. Biến giải thích cho việc trả lời câu hỏi:**
* Giới tính (gender)
* Tuổi (age)
* Tình trạng huyết áp (hypertension)
* Bệnh tim (heart_disease)
* Tình trạng kết hôn (ever_married)
* Loại công việc (work_type)
* Môi trường sống (Residence_type)
* Lượng đường trung bình trong máu (avg_glucose_level)
* Tình trạng hút thuốc (smoking_status)

**2. Biến kết quả:**
* Tình trạng đột quỵ (stroke)

**3. Các nhóm so sánh:**
* Nhóm những người có bệnh tim
* Nhóm những người cao huyết áp
* Nhóm những người có lượng đường trong máu cao

**4. Phân tích dữ liệu:**

```python
import matplotlib.pyplot as plt

smoking = data['smoking_status']
stroke = data['stroke']
heart_disease = data['heart_disease']
hypertension = data['hypertension']
avg_glucose_level = data['avg_glucose_level']
age = data['age']


count_smoke = 0; #biến đếm số người hút thuốc
count_smoke_stroke = 0 #biến đếm số người hút thuốc bị đột quỵ
count_nvsmoke = 0 #biến đếm số người không bao giờ hút thuốc
count_nvsmoke_stroke = 0 #biến đếm số người không bao giờ hút thuốc nhưng bị đột quỵ
count_unk = 0 #biến đếm số người không rõ có hút thuốc hay không 
count_unk_stroke = 0 #biến đếm số người không rõ có hút thuốc hay không nhưng bị đột quỵ

count_heart = 0
count_unheart = 0

count_hypertension = 0
count_unhypertension = 0


for i in range(len(smoking)):
    #Có hút thuốc
    if(smoking[i] != "Unknown" and smoking[i] != "never smoked"):
        count_smoke += 1
    #Có hút thuốc và bị đột quỵ
    if(smoking[i] != "Unknown" and smoking[i] != "never smoked" and stroke[i] == 1):
        count_smoke_stroke += 1
    
    #Không rõ có hút thuốc hay không
    if(smoking[i] == "Unknown"):
        count_unk += 1
    #không rõ có hút thuốc hay không nhưng bị đột quỵ
    if(smoking[i] == "Unknown" and stroke[i] == 1):
        count_unk_stroke += 1
    
    #Không bao giờ hút thuốc
    if(smoking[i] == "never smoked"):
        count_nvsmoke += 1
    #Không bao giờ hút thuốc nhưng bị đột quỵ
    if(smoking[i] == "never smoked" and stroke[i] == 1):
        count_nvsmoke_stroke += 1

for i in range(len(heart_disease)):
    #Có vấn đề về tim mạch
    if heart_disease[i] == 1:
        count_heart += 1
    else:
        count_unheart += 1

for i in range(len(hypertension)):
    #Có vấn đề về huyết áp
    if hypertension[i] == 1:
        count_hypertension += 1
    else:
        count_unhypertension += 1
```

```python
print("Số người hút thuốc: ", count_smoke)
print("Số người hút thuốc và bị đột quỵ: ", count_smoke_stroke)

print("Số người không rõ có hút thuốc không: ", count_unk)
print("Số người không rõ có hút thuốc không nhưng bị đột quỵ: ", count_unk_stroke)

print("Số người không bao giờ hút thuốc: ", count_nvsmoke)
print("Số người không bao giờ hút thuốc nhưng bị đột quỵ: ", count_nvsmoke_stroke)
```

* Một số tóm tắt thống kê:
> Khảo sát 5110 quan sát có 1674 người hút thuốc, 1892 không hút thuốc. 

> Trong 1674 người hút thuốc có 112 người bị đột quỵ chiếm 6.69%. Trong 1892 người không hút thuốc có 90 người đột quỵ chiếm 4.76%

* Biểu đồ trực quan

```python
figure, ax = plt.subplots()
ax.axis('equal')
cont = ['Có vấn đề tim mạch', 'Không có vấn đề tim mạch']
so_lieu = [count_heart, count_unheart]

ax.pie(so_lieu, labels = cont, autopct='%1.2f%%')
ax.set(title = 'Biểu đồ 1. Tình trạng về tim mạch trong 5110 quan sát')
plt.show()
```

```python
figure, ax = plt.subplots()
ax.axis('equal')
cont = ['Có vấn đề huyết áp', 'Không có vấn đề huyết áp']
so_lieu = [count_hypertension, count_unhypertension]

ax.pie(so_lieu, labels = cont, autopct='%1.2f%%')
ax.set(title = 'Biểu đồ 2. Tình trạng về huyết áp trong 5110 quan sát')
plt.show()
```

```python
fig, ax = plt.subplots()
x = avg_glucose_level
y = age
plt.scatter(x,y)
plt.title('Biểu đồ 3. Mối quan hệ giữa độ tuổi và lượng đường trung bình trong máu')
plt.xlabel('avg_glucose_level')
plt.ylabel('age')
```

```python
figure, ax = plt.subplots()
ax.axis('equal')
cont = ['smoked', 'never smoked', 'unknow']
so_lieu = [count_smoke, count_nvsmoke, count_unk]

ax.pie(so_lieu, labels = cont, autopct='%1.2f%%')
ax.set(title = 'Biểu đồ 4. Tình trạng hút thuốc trong 5110 quan sát')
plt.show()
```

Từ các biểu đồ trực quan hóa trên, chúng ta có một số nhận xét như sau:
1. Biểu đồ 1 và biểu đồ 2 cho thấy trong các trường hợp được khảo sát, số người có vấn đề về tim mạch hoặc huyết áp chiếm phần khá nhỏ.
2. Theo một trang web uy tín[3] thì:

    > A blood sugar level less than 140 mg/dL (7.8 mmol/L) is normal.

    Nhìn vào biểu đồ 3, tuy số tuổi được khảo sát phân bố đều từ trẻ em cho đến người lớn tuổi nhưng lượng đường trung bình trong máu của những được khảo sát đa phần đều bình thường (<140 mg/dL).
3. Biểu đồ 4 cho chúng ta thấy về tình trạng hút thuốc lá ở những người được khảo sát. Ta thấy rằng số người hút thuốc vẫn còn chiếm một phần lớn trên tổng thể và tỉ lệ mắc bệnh của những người này cao hơn những người không hút thuốc. (Cụ thể: Tỉ lệ giữ người hút thuốc bị đột quỵ và số người không hút thuốc nhưng bị đột quỵ chiếm là 1.41, *được dẫn từ: Một số tóm tắt thống kê*).
